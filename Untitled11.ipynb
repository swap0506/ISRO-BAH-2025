```python
# =========================
# 1. CLEAN COLUMN NAMES
# =========================
df.columns = df.columns.str.strip()


# =========================
# 2. KEEP ONLY REQUIRED COVERAGE COLUMNS
# =========================
coverage_keep_cols = [
    'coverage.expectedCount',
    'coverage.expectedInterval',
    'coverage.observedCount',
    'coverage.observedInterval',
    'coverage.percentComplete',
    'coverage.percentCoverage'
]

# Get all coverage columns
coverage_cols = [col for col in df.columns if col.startswith('coverage.')]

# Find columns to drop
drop_cols = [col for col in coverage_cols if col not in coverage_keep_cols]

# Drop them safely
df.drop(columns=drop_cols, inplace=True, errors='ignore')

print(f"Dropped {len(drop_cols)} coverage columns")


# =========================
# 3. CHECK REMAINING COVERAGE COLUMNS
# =========================
for col in coverage_keep_cols:
    if col in df.columns:
        print(f"\n{col}:")
        print(df[col].head())


# =========================
# 4. PARAMETER → UNIT MAPPING (UNIQUE)
# =========================
param_unit_map = df[['parameter.name', 'parameter.units']].drop_duplicates()

print("\nParameter → Units mapping:")
print(param_unit_map)


# =========================
# 5. CONVERT parameter.name → COLUMNS (IMPORTANT)
# =========================
df_wide = df.pivot_table(
    index='period.datetimeFrom.utc', # adjust if needed
    columns='parameter.name',
    values='value',
    aggfunc='mean'
).reset_index()


```

```python
# remove column name hierarchy
df_wide.columns.name = None

# clean column names FIRST
df_wide.columns = [str(col).strip().lower() for col in df_wide.columns]

# THEN use column
df_wide['date'] = pd.to_datetime(
    df_wide['period.datetimefrom.utc'],
    errors='coerce'
)

df_wide = df_wide.sort_values('date')
```

```python
print("\nFinal transformed data:")
print(df_wide.head())
print("Shape:", df_wide.shape)
```

```python
# =========================
# MERGE MET DATA (ADD HERE)
# =========================

# make column names same
df.rename(columns={'period.datetimeFrom.utc': 'period.datetimefrom.utc'}, inplace=True)

met_cols = ['AOD', 'T2M', 'RH2M', 'U2M', 'V2M', 'PS']
met_cols = [c for c in met_cols if c in df.columns]

if len(met_cols) > 0:
    meta_df = df[['period.datetimefrom.utc'] + met_cols].drop_duplicates()

    df_wide = pd.merge(
        df_wide,
        meta_df,
        on='period.datetimefrom.utc',
        how='left'
    )
```

```python




# =========================
# 5. FEATURE ENGINEERING
# =========================

# wind speed
if 'U2M' in df_wide.columns and 'V2M' in df_wide.columns:
    df_wide['wind_speed'] = np.sqrt(df_wide['U2M']**2 + df_wide['V2M']**2)

# temperature rename
if 'T2M' in df_wide.columns:
    df_wide.rename(columns={'T2M': 'temperature'}, inplace=True)

if 'RH2M' in df_wide.columns:
    df_wide.rename(columns={'RH2M': 'relativehumidity'}, inplace=True)

# =========================
# 6. CLEAN DATA
# =========================
df_wide.replace([-999, -199], np.nan, inplace=True)

# fill missing (VERY IMPORTANT)
df_wide = df_wide.ffill().bfill()

# =========================
# 7. TIME FEATURES
# =========================
df_wide['hour'] = df_wide['date'].dt.hour
df_wide['day'] = df_wide['date'].dt.day
df_wide['month'] = df_wide['date'].dt.month

# =========================
# 8. LAG + ROLLING
# =========================
df_wide['pm25_lag1'] = df_wide['pm25'].shift(1)
df_wide['pm25_lag2'] = df_wide['pm25'].shift(2)

df_wide['pm25_roll3'] = df_wide['pm25'].rolling(3).mean()
df_wide['pm25_roll6'] = df_wide['pm25'].rolling(6).mean()

# fill again after lag
df_wide = df_wide.ffill().bfill()

# =========================
# 9. SELECT FEATURES
# =========================
keep_cols = [
    'pm25',
    'pm10', 'so2', 'no2', 'co', 'o3',
    'temperature', 'relativehumidity',
    'wind_speed', 'PS',
    'AOD',
    'hour', 'day', 'month',
    'pm25_lag1', 'pm25_lag2',
    'pm25_roll3', 'pm25_roll6'
]

df_model = df_wide[[c for c in keep_cols if c in df_wide.columns]]

# numeric only
df_model = df_model.select_dtypes(include=['number'])

# =========================
# 10. CREATE X, y
# =========================
X = df_model.drop(columns=['pm25'])
y = df_model['pm25']

print("Final dataset shape:", X.shape)

# =========================
# 11. SPLIT
# =========================
split = int(len(X) * 0.8)

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# =========================
# 12. MODELS
# =========================
def evaluate(model, X_train, X_test, y_train, y_test, name):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    print(f"\n{name}")
    print("RMSE:", np.sqrt(mean_squared_error(y_test, pred)))
    print("MAE :", mean_absolute_error(y_test, pred))
    print("R2  :", r2_score(y_test, pred))


xgb = XGBRegressor(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

rf = RandomForestRegressor(n_estimators=200, random_state=42)
gbr = GradientBoostingRegressor()

stack = StackingRegressor(
    estimators=[("xgb", xgb), ("rf", rf), ("gbr", gbr)],
    final_estimator=Ridge(),
    cv=KFold(n_splits=3)
)

xgb_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
    ("model", xgb)
])

stack_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
    ("scale", StandardScaler()),
    ("model", stack)
])

# =========================
# 13. TIME SERIES CV
# =========================
tscv = TimeSeriesSplit(n_splits=5)

cv_xgb = cross_val_score(
    xgb_pipe, X, y,
    scoring="neg_root_mean_squared_error",
    cv=tscv
).mean()

cv_stack = cross_val_score(
    stack_pipe, X, y,
    scoring="neg_root_mean_squared_error",
    cv=tscv
).mean()

print("\nCross Validation RMSE:")
print("XGBoost :", -cv_xgb)
print("Stacking:", -cv_stack)

# =========================
# 14. FINAL EVALUATION
# =========================
evaluate(xgb_pipe, X_train, X_test, y_train, y_test, "XGBoost")
evaluate(stack_pipe, X_train, X_test, y_train, y_test, "Stacking")

# =========================
# 15. FEATURE IMPORTANCE
# =========================
xgb_pipe.fit(X_train, y_train)

importances = xgb_pipe.named_steps["model"].feature_importances_
feat_imp = pd.Series(importances, index=X.columns).sort_values(ascending=False)

print("\nTop Features:")
print(feat_imp.head(10))
```
